В даному випадку отримаємо **варіант №2** - **Дослідження втрат танків**:
1. Створіть функцію, яка приймає назву файлу та назву стовпця як параметр і повертає масив **numpy** зі значеннями стовпця.
2. Використовуючи функцію з попереднього завдання, створіть масив **numpy** зі значень стовпця "tank".
3. Знайдіть добові втрати танків.
4. Знайдіть 4 найбільші добові втрати танків.
5. Визначте, скільки танків було знищено навесні 2023 року.
6. Знайдіть середнє значення знищених танків за останні 100 днів війни.
7. Створіть графік втрат танків за минулий рік (починаючи з останньої дати в наборі даних).

```python
import csv
import numpy as np

def get_csv_column_data(file_name, column_name):
    """
    Функція для отримання даних стовпця з файлу CSV та їх перетворення в масив NumPy.

    Параметри:
    file_name (str): Назва файлу CSV, з якого потрібно отримати дані стовпця.
    column_name (str): Назва стовпця, дані якого потрібно отримати.

    Повертає:
    numpy.ndarray: Масив NumPy, що містить значення обраного стовпця.
    None: Якщо обранного файлу або стовпця не існує.
    """
    # Створюємо список для зберігання даних стовпця
    column_data = []
    
    try:
        with open(file_name, 'r') as file:
            csv_reader = csv.reader(file)
            # Отримуємо рядок заголовків, тим самим пропускаємо його
            header = next(csv_reader)
    
            try:
                # Знаходимо індекс потрібного стовпця за назвою
                column_index = header.index(column_name)
            except ValueError:
                print(f"Стовпець '{column_name}' не знайдено у файлі '{file_name}'.")
                return None

            # Заповнюємо список даними зі стовпця
            for row in csv_reader:
                column_data.append(row[column_index])

    except FileNotFoundError:
        print(f"Помилка: Файл '{file_name}' не існує.")
        return None
    except Exception as e:
        print(f"Помилка: {e}")
        return None

    # Перетворюємо список в масив NumPy
    column_data = np.array(column_data)
    return column_data
```

```python
def get_daily_tank_losses(tank_losses_data):
    """
    Функція для обчислення добових втрат танків.

    Параметри:
    tank_losses_data (numpy.ndarray): Масив NumPy, що містить загальні дані про втрати танків.

    Повертає:
    numpy.ndarray: Масив NumPy, що містить добові втрати танків.
    """
    daily_tank_losses = np.diff(tank_losses_data.astype(int))
    return daily_tank_losses
```

```python
def get_four_largest_daily_tank_losses(daily_tank_losses):
    """
    Функція для знаходження чотирьох найбільших добових втрат танків.

    Параметри:
    daily_tank_losses (numpy.ndarray): Масив NumPy, що містить добові втрати танків.

    Повертає:
    numpy.ndarray: Масив NumPy, що містить чотири найбільші добові втрати танків.
    """
    unique_values = np.unique(daily_tank_losses)
    four_largest_daily_tank_losses = np.sort(unique_values)[-4:][::-1]
    return four_largest_daily_tank_losses
```

```python
def get_spring2023_losses(tank_losses_data, losses_dates):
    """
    Функція для знаходження кількості знищених танків за весну 2023 року.

    Параметри:
    tank_losses_data (numpy.ndarray): Масив NumPy, що містить загальні дані про втрати танків.
    losses_dates (numpy.ndarray): Масив NumPy, що містить дати втрат.

    Повертає:
    int: Кількість танків, які були знищенні за весну 2023 року.
    """

    # Знаходимо індекси початку та кінця весни 2023 року
    start_index = np.where(losses_dates == '2023-03-01')[0][0]
    end_index = np.where(losses_dates == '2023-05-31')[0][0]
    
    # Обчислення втрат танків для весни 2023 року
    last_spring_losses_number = tank_losses_data.astype(int)[end_index]
    first_spring_losses_number = tank_losses_data.astype(int)[start_index]
    spring2023_losses = np.subtract(last_spring_losses_number, first_spring_losses_number)
    return spring2023_losses
```

```python
def get_average_100days_losses(daily_tank_losses):
    """
    Функція для знаходження середнього значення знищених танків за останні 100 днів війни.

    Параметри:
    daily_tank_losses (numpy.ndarray): Масив NumPy, що містить добові втрати танків.

    Повертає:
    float: Середнє значення знищених танків за останні 100 днів війни.
    """
    last_100days = daily_tank_losses[-100:]
    average_100days_losses = round(np.mean(last_100days))
    return average_100days_losses
```

```python
import matplotlib.pyplot as plt

def create_tank_losses_plot(tank_losses_data):
    last_year_losses = tank_losses_data[-365:]
    plt.figure(figsize=(8, 8), dpi=100)
    plt.plot(last_year_losses.astype(int), label='Втрати танків', linestyle=':', color='red', linewidth=2)
    plt.title('Втрати танків за останній рік')
    plt.xlabel('Дні війни')
    plt.ylabel('Втрати танків')
    plt.grid(True, linestyle='--', linewidth=1)
    plt.legend()

    plt.savefig('tank_losses_last_year.png')
    plt.show()
```

```python
tank_losses_data = get_csv_column_data('russia_losses_equipment.csv', 'tank')
daily_tank_losses = get_daily_tank_losses(tank_losses_data)
four_largest_daily_tank_losses = get_four_largest_daily_tank_losses(daily_tank_losses)

losses_dates = get_csv_column_data('russia_losses_equipment.csv', 'date')
spring2023_losses = get_spring2023_losses(tank_losses_data, losses_dates)
average_100days_losses = get_average_100days_losses(daily_tank_losses)

print('Добові втрати танків (останні 7 днів):', daily_tank_losses[-7:])
print('Чотири найбільші добові втрати танків:', four_largest_daily_tank_losses)
print('Танків було знищено навесні 2023 року:', spring2023_losses)
print('Середнє значення знищених танків за останні 100 днів війни:', average_100days_losses)

create_tank_losses_plot(tank_losses_data) 
```
