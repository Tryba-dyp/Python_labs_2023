# Алгоритм

```python
def find_squares_sum(N):
    for a in range(1, int(N ** 0.5) + 1):
        b_square = N - a**2
        b = int(b_square**0.5)
        
        if b_square == b**2:
            return a, b
    return None
```

```python
N = 100
result = find_squares_sum(N)

if result is not None:
    a, b = result
    print(f"{a}^2 + {b}^2 = {N}")
else:
    print(f"Число {N} не можна представити квадратами цілих чисел.")
```
